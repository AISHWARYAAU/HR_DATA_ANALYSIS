## Step 6: Data Visualisation

```python
print(data.columns)

# To create a Dataframe object to hold the data in a tabular format
df = pd.DataFrame(data.head(20))
```

```python
# Generate a pie chart to visualize the distribution of values in the 'Edu' column
df['EducationField'].value_counts().plot(kind='pie', autopct='%1.1f%%')

# Add a title to the pie chart indicating the variable being visualized
plt.title("EducationField")

# Display the pie chart
plt.show()
```

```python
# Generate a bar chart to visualize the distribution of values in the 'DailyRate' column
df['DailyRate'].value_counts().plot(kind='bar', color='green')

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('Number of DailyRate ')

# Add a label to the y-axis indicating the count of each value
plt.ylabel('Count')

# Add a title to the bar chart indicating the purpose of the visualization
plt.title('Distribution of Number of DailyRate')

# Display the bar chart
plt.show()
```

```python
# Generate a horizontal bar chart to visualize the distribution of values in the 'Department' column
df['Department'].value_counts().plot(kind='barh', color='yellow')

# Add a label to the x-axis indicating the count of each value
plt.xlabel('Count')

# Add a label to the y-axis indicating the variable being visualized
plt.ylabel('Department')

# Add a title to the horizontal bar chart indicating the purpose of the visualization
plt.title('Distribution of Department Levels')

# Display the horizontal bar chart
plt.show()
```

```python
# Generate a histogram to visualize the distribution of values in the 'Income' column
plt.hist(df['MonthlyIncome'], bins=20, color='green', edgecolor='black')

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('Income')

# Add a label to the y-axis indicating the frequency of observations
plt.ylabel('Frequency')

# Add a title to the histogram indicating the purpose of the visualization
plt.title('Distribution of Income')

# Add grid lines to the histogram for better visualization
plt.grid(True)

# Display the histogram
plt.show()
```

```python

plt.scatter(df['MonthlyIncome'], df['JobInvolvement'], color='blue', alpha=0.5)

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('Income')

# Add a label to the y-axis indicating the variable being visualized
plt.ylabel('Number of JobInvolvement')

# Add a title to the scatter plot indicating the purpose of the visualization
plt.title('Scatter Plot: Income vs. Number of JobInvolvement')

# Add grid lines to the scatter plot for better visualization
plt.grid(True)

# Display the scatter plot
plt.show()
```
